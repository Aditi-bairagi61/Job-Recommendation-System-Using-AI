# Clustering the Job Data using csv created previously

## Also providing skills as an input

### Visualising individual Clusters individually

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics.pairwise import cosine_similarity

# Load the CSV file generated earlier
df = pd.read_csv("all_skills_matrix.csv", index_col=0)

# Define the number of clusters based on experimentation or domain knowledge (adjustable)
num_clusters = 21  # Adjust based on the dataset size
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(df.iloc[:, :-1])

# Function to recommend jobs based on input skills
def recommend_jobs(input_skills):
    # Create a binary vector for input skills matching the columns of the DataFrame
    input_vector = [1 if skill in input_skills else 0 for skill in df.columns[:-1]]
    
    # Compute similarity scores between the input vector and all job rows
    similarity_scores = cosine_similarity([input_vector], df.iloc[:, :-1])[0]
    
    # Get jobs in order of relevance
    sorted_job_indices = similarity_scores.argsort()[::-1]
    sorted_jobs = df.index[sorted_job_indices]
    
    # Print jobs in order of relevance
    print("Job recommendations based on input skills (sorted by relevance):")
    for i, job in enumerate(sorted_jobs, 1):
        print(f"{i}. {job} (Similarity score: {similarity_scores[sorted_job_indices[i-1]]:.2f})")

# Example input skills
input_skills = ["Policy development",
            "Problem-solving skills",
            "Communication skills",
            "Budgeting and financial management",
            "Project management expertise",]
recommend_jobs(input_skills)

# Dimensionality Reduction for Visualization (PCA and t-SNE)
# Step 1: Apply PCA to reduce dimensionality to 50 components (faster computation for t-SNE)
pca = PCA(n_components=50)
pca_result = pca.fit_transform(df.iloc[:, :-1])

# Step 2: Apply t-SNE to reduce to 2 dimensions for visualization
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(pca_result)

# Add t-SNE results to the DataFrame
df['tsne_1'] = tsne_result[:, 0]
df['tsne_2'] = tsne_result[:, 1]

# Plotting the overall clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x="tsne_1", y="tsne_2",
    hue="Cluster",
    palette=sns.color_palette("hsv", num_clusters),
    data=df,
    legend="full",
    s=60
)
plt.title("Visualization of Job Clusters")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.legend(title="Cluster")
plt.show()

# Display and save data points for each cluster, along with top skills
for cluster in range(num_clusters):
    cluster_data = df[df['Cluster'] == cluster]
    print(f"\nCluster {cluster} contains the following jobs:")
    print(f"Number of data points in Cluster {cluster}: {len(cluster_data)}")
    print(cluster_data.index.tolist())
    
    # Calculate mean of each skill for the cluster to determine dominant skills
    skill_means = cluster_data.iloc[:, :-3].mean().sort_values(ascending=False)  # Exclude non-skill columns
    top_skills = skill_means[skill_means > 0.02]  # Threshold for dominant skills (adjustable)

    print("Top skills for this cluster:")
    print(top_skills)
    
    # Plot data points for the current cluster
    plt.figure(figsize=(8, 6))
    sns.scatterplot(
        x="tsne_1", y="tsne_2",
        data=cluster_data,
        color=sns.color_palette("hsv", num_clusters)[cluster],
        s=60
    )
    plt.title(f"Visualization of Cluster {cluster}")
    plt.xlabel("t-SNE Dimension 1")
    plt.ylabel("t-SNE Dimension 2")
    plt.show()  # or save using plt.savefig(f"cluster_{cluster}.png")
```
