```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Load the binary matrix from the CSV file
binary_matrix = pd.read_csv("jov_vs_salary_binary_matrix.csv", index_col="Job Title")

# Function to search by city
def search_by_city(city, binary_matrix):
    """
    Returns a list of job titles available in a specified city.
    """
    if city in binary_matrix.columns:
        jobs_in_city = binary_matrix[binary_matrix[city] == 1].index.tolist()
        return jobs_in_city if jobs_in_city else "No jobs available in this city."
    else:
        return "City not found in the database."

# Function to search by job title
def search_by_job_title_salary(job_title, binary_matrix):
    """
    Returns a list of cities where the specified job title is available.
    """
    if job_title in binary_matrix.index:
        cities_for_job = binary_matrix.loc[job_title][binary_matrix.loc[job_title] == 1].index.tolist()
        return cities_for_job if cities_for_job else "This job is not available in any city."
    else:
        return "Job title not found in the database."

# Function to perform clustering on cities
def cluster_cities(binary_matrix, num_clusters=3):
    """
    Clusters cities based on job availability using KMeans and returns clusters of cities.
    """
    # Transpose the matrix so that cities are rows and job titles are columns
    city_job_matrix = binary_matrix.T

    # Apply K-means clustering on the cities
    kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(city_job_matrix)

    # Add cluster labels to the DataFrame
    city_job_matrix['Cluster'] = kmeans.labels_

    # Display the clustered cities
    city_clusters = {}
    for cluster in range(num_clusters):
        cities_in_cluster = city_job_matrix[city_job_matrix['Cluster'] == cluster].index.tolist()
        city_clusters[cluster] = cities_in_cluster
        print(f"Cities in Cluster {cluster}: {cities_in_cluster}")
    
    # Visualizing city clusters
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=city_job_matrix, x=city_job_matrix.index, y=[0]*len(city_job_matrix), hue='Cluster', palette='viridis')
    plt.title("City Clusters Based on Job Availability")
    plt.xlabel("Cities")
    plt.ylabel("Cluster")
    plt.xticks(rotation=90)
    plt.show()
    
    return city_clusters

# Function to perform clustering on job titles
def cluster_jobs(binary_matrix, num_clusters=3):
    """
    Clusters job titles based on city availability using KMeans and returns clusters of job titles.
    """
    # Apply K-means clustering on the job titles
    kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(binary_matrix)

    # Add cluster labels to the DataFrame
    binary_matrix['Cluster'] = kmeans.labels_

    # Display the clustered job titles
    job_clusters = {}
    for cluster in range(num_clusters):
        jobs_in_cluster = binary_matrix[binary_matrix['Cluster'] == cluster].index.tolist()
        job_clusters[cluster] = jobs_in_cluster
        print(f"Jobs in Cluster {cluster}: {jobs_in_cluster}")
    
    # Visualizing job clusters
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=binary_matrix, x=binary_matrix.index, y=[0]*len(binary_matrix), hue='Cluster', palette='viridis')
    plt.title("Job Clusters Based on City Availability")
    plt.xlabel("Job Titles")
    plt.ylabel("Cluster")
    plt.xticks(rotation=90)
    plt.show()
    
    return job_clusters

# Example Usage:
# # 1. Search by city
# city = "Los Angeles, CA"
# jobs = search_by_city(city, binary_matrix)
# print(f"Jobs available in {city}: {jobs}")

# 2. Search by job title
job_title = "Facilities Administrator"
salary = search_by_job_title_salary(job_title, binary_matrix)
print(f"Avg. Salary of {job_title}: {salary}")

# # 3. Cluster salary by job availability
# print("\nClustering salary by job availability:")
# city_clusters = cluster_cities(binary_matrix, num_clusters=3)

# # 4. Cluster job titles by city availability
# print("\nClustering job titles by city availability:")
# job_clusters = cluster_jobs(binary_matrix, num_clusters=3)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics.pairwise import cosine_similarity

# Load the CSV file generated earlier
df = pd.read_csv("all_skills_matrix.csv", index_col=0)

# Define the number of clusters based on experimentation or domain knowledge (adjustable)
num_clusters = 21  # Adjust based on the dataset size
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(df.iloc[:, :-1])

# Function to recommend jobs based on input skills
def recommend_jobs(input_skills):
    # Create a binary vector for input skills matching the columns of the DataFrame
    input_vector = [1 if skill in input_skills else 0 for skill in df.columns[:-1]]
    
    # Compute similarity scores between the input vector and all job rows
    similarity_scores = cosine_similarity([input_vector], df.iloc[:, :-1])[0]
    
    # Get jobs in order of relevance
    sorted_job_indices = similarity_scores.argsort()[::-1]
    sorted_jobs = df.index[sorted_job_indices]
    
    # Print jobs in order of relevance
    print("Job recommendations based on input skills (sorted by relevance):")
    for i, job in enumerate(sorted_jobs, 1):
        print(f"{i}. {job} (Similarity score: {similarity_scores[sorted_job_indices[i-1]]:.2f})")
        cities = search_by_job_title_location(job, binary_matrix)
        print(f"Cities offering {job}: {cities}")
        salary = search_by_job_title_salary(job_title, binary_matrix)
        print(f"Avg. Salary of {job_title}: {salary}")

# Example input skills
input_skills = ["Policy development",
            "Problem-solving skills",
            "Communication skills",
            "Budgeting and financial management",
            "Project management expertise",]
recommend_jobs(input_skills)

# Dimensionality Reduction for Visualization (PCA and t-SNE)
# Step 1: Apply PCA to reduce dimensionality to 50 components (faster computation for t-SNE)
pca = PCA(n_components=50)
pca_result = pca.fit_transform(df.iloc[:, :-1])

# Step 2: Apply t-SNE to reduce to 2 dimensions for visualization
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(pca_result)

# Add t-SNE results to the DataFrame
df['tsne_1'] = tsne_result[:, 0]
df['tsne_2'] = tsne_result[:, 1]

# Plotting the overall clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x="tsne_1", y="tsne_2",
    hue="Cluster",
    palette=sns.color_palette("hsv", num_clusters),
    data=df,
    legend="full",
    s=60
)
plt.title("Visualization of Job Clusters")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.legend(title="Cluster")
plt.show()

# Display and save data points for each cluster, along with top skills
for cluster in range(num_clusters):
    cluster_data = df[df['Cluster'] == cluster]
    print(f"\nCluster {cluster} contains the following jobs:")
    print(f"Number of data points in Cluster {cluster}: {len(cluster_data)}")
    print(cluster_data.index.tolist())
    
    # Calculate mean of each skill for the cluster to determine dominant skills
    skill_means = cluster_data.iloc[:, :-3].mean().sort_values(ascending=False)  # Exclude non-skill columns
    top_skills = skill_means[skill_means >= 0.2]  # Threshold for dominant skills (adjustable)

    print("Top skills for this cluster:")
    print(top_skills)
    
    # Plot data points for the current cluster
    plt.figure(figsize=(8, 6))
    sns.scatterplot(
        x="tsne_1", y="tsne_2",
        data=cluster_data,
        color=sns.color_palette("hsv", num_clusters)[cluster],
        s=60
    )
    plt.title(f"Visualization of Cluster {cluster}")
    plt.xlabel("t-SNE Dimension 1")
    plt.ylabel("t-SNE Dimension 2")
    plt.show()  # or save using plt.savefig(f"cluster_{cluster}.png")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics.pairwise import cosine_similarity

# Load the CSV file generated earlier
df = pd.read_csv("all_skills_matrix.csv", index_col=0)

# Load the binary matrix for cities and salaries
binary_matrix = pd.read_csv("jov_vs_salary_binary_matrix.csv", index_col="Job Title")

# Define the number of clusters based on experimentation or domain knowledge (adjustable)
num_clusters = 21  # Adjust based on the dataset size
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(df.iloc[:, :-1])

# Function to search by city
def search_by_job_title_location(job_title, binary_matrix):
    if job_title in binary_matrix.index:
        cities = binary_matrix.loc[job_title][binary_matrix.loc[job_title] == 1].index.tolist()
        return cities if cities else "This job is not available in any city."
    else:
        return "Job title not found in the database."

# Function to search by salary
def search_by_job_title_salary(job_title, binary_matrix):
    if job_title in binary_matrix.index:
        # Assuming salary is stored in a specific column, e.g., 'Salary'
        salary_col = [col for col in binary_matrix.columns if "salary" in col.lower()]
        if salary_col:
            return binary_matrix.loc[job_title, salary_col[0]]
        else:
            return "Salary data not available."
    else:
        return "Job title not found in the database."

# Function to recommend jobs based on input skills
def recommend_jobs(input_skills):
    # Create a binary vector for input skills matching the columns of the DataFrame
    input_vector = [1 if skill in input_skills else 0 for skill in df.columns[:-1]]
    
    # Compute similarity scores between the input vector and all job rows
    similarity_scores = cosine_similarity([input_vector], df.iloc[:, :-1])[0]
    
    # Get jobs in order of relevance
    sorted_job_indices = similarity_scores.argsort()[::-1]
    sorted_jobs = df.index[sorted_job_indices]
    
    # Print jobs in order of relevance
    print("Job recommendations based on input skills (sorted by relevance):")
    for i, job in enumerate(sorted_jobs, 1):
        print(f"{i}. {job} (Similarity score: {similarity_scores[sorted_job_indices[i-1]]:.2f})")
        cities = search_by_job_title_location(job, binary_matrix)
        print(f"Cities offering {job}: {cities}")
        salary = search_by_job_title_salary(job, binary_matrix)
        print(f"Avg. Salary of {job}: {salary}")

# Example input skills
input_skills = [
    "Policy development",
    "Problem-solving skills",
    "Communication skills",
    "Budgeting and financial management",
    "Project management expertise",
]
recommend_jobs(input_skills)

# Dimensionality Reduction for Visualization (PCA and t-SNE)
# Step 1: Apply PCA to reduce dimensionality to 50 components (faster computation for t-SNE)
pca = PCA(n_components=50)
pca_result = pca.fit_transform(df.iloc[:, :-1])

# Step 2: Apply t-SNE to reduce to 2 dimensions for visualization
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(pca_result)

# Add t-SNE results to the DataFrame
df['tsne_1'] = tsne_result[:, 0]
df['tsne_2'] = tsne_result[:, 1]

# Plotting the overall clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x="tsne_1", y="tsne_2",
    hue="Cluster",
    palette=sns.color_palette("hsv", num_clusters),
    data=df,
    legend="full",
    s=60
)
plt.title("Visualization of Job Clusters")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.legend(title="Cluster")
plt.show()
```
